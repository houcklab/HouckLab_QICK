```python
def fit(params, x):
    A = params['A']
    B = params['B']
    return A/x + B

def error_func(params, xs, ys):
    return ys - fit(params, xs)

def analyzeResistances(widths, voltagesMV, label = '', scalingFactor = 10000, desiredJuncResistance=10000):
    print('############################################')
    print(label)
    
    assert len(widths) == len(voltagesMV)
    widthsClean = []
    resistances = []
    for ii in range(len(widths)):
        if len(voltagesMV[ii]) == 0:
            continue
        elif len(voltagesMV[ii]) == 1:
            res = voltagesMV[ii][0]*scalingFactor
        elif len(voltagesMV[ii]) == 2:
            res = (voltagesMV[ii][1] - voltagesMV[ii][0])*scalingFactor
        elif len(voltagesMV[ii]) == 3:
            res = ((voltagesMV[ii][1] - voltagesMV[ii][0]) + (voltagesMV[ii][2] - voltagesMV[ii][1]))/2 * scalingFactor
        else:
            assert False
        if res > 500:
            widthsClean.append(widths[ii])
            resistances.append(res)
    numIgnored = len(widths) - len(widthsClean)
    print('Number of widths ignored: ' + str(numIgnored))
    print('Approximate yield: ' + str((len(widths) - numIgnored)/len(widths)))

    init_params = Parameters()
    init_params.add('A', value=1e6)
    init_params.add('B', value=0)
    
    fitOut = minimize(error_func, init_params, args = (widthsClean, resistances))
    
    print()
    print('Fit Parameters')
    fitOut.params.pretty_print()
    print()
    print('Resistances in kOhm')
    print(np.array(resistances)/1000)
    
    widthsCleanInterp = np.linspace(min(widthsClean), max(widthsClean), 100)

    desiredWidth = widthsCleanInterp[np.argmin(np.abs(desiredJuncResistance - fit(fitOut.params, widthsCleanInterp)))]
    desiredSideLength = desiredWidth
    print()
    print('If you want a junction with resistance (ohms): ')
    print(desiredJuncResistance)
    print('Then make a junction with width: ')
    print(desiredWidth)
    
    fig, ax = plt.subplots()
    ax.plot(widthsClean, resistances, 'o')
    ax.plot(widthsCleanInterp, fit(fitOut.params, widthsCleanInterp))
#     ax.plot([desiredArea, desiredArea], [min(resistances), max(resistances)])
    ax.set_xlabel('Width (um)')
    ax.set_ylabel('Resistance (Ohms)')
    ax.set_title(label)
    plt.show()
```
