```python
import numpy as np
import matplotlib.pyplot as plt

ellB = 1.0  # magnetic length
j_values = [0, 1]

# create grid
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)
R2 = X**2 + Y**2

fig, axes = plt.subplots(1, 2, figsize=(10,5))

for ax, j in zip(axes, j_values):
    psi2 = (R2**j) * np.exp(-R2/(2*ellB**2))
    ax.contourf(X, Y, psi2, levels=50, cmap='viridis')
    ax.set_title(f'|Ïˆ_0{j}|^2')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```
